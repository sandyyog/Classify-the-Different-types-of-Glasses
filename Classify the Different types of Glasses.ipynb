```python
import pandas as pd
import numpy as np
df=pd.read_csv('glass.csv')
df
```

```python
x=df.iloc[:,0:9]
x
```

```python
y=df.iloc[:,9:10]
y
```

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
ss_out=ss.fit_transform(x)
x=ss_out
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

```python
from sklearn.tree import DecisionTreeClassifier
obj=DecisionTreeClassifier(criterion='entropy')
output=obj.fit(x_train,y_train)
output1=output.predict(x_test)
output1
```

```python
from sklearn.model_selection import cross_val_score
z=cross_val_score(estimator=obj,X=x_train,y=y_train,cv=5)
z
```

```python
from sklearn.ensemble import RandomForestClassifier
fc=RandomForestClassifier(n_estimators=2)
a=fc.fit(x_train,y_train)
b=a.predict(x_test)
b
```

```python
x_train=pd.DataFrame(x_train)
x_train
```

```python
from sklearn.model_selection import GridSearchCV
a= GridSearchCV(estimator=fc,param_grid={'n_estimators':[1,2,4,6,8,10]},scoring='accuracy',cv= 5)
a.fit(x_train,y_train)
a.best_params_
```
